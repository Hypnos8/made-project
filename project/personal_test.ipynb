```python
import re
pd.set_option('display.max_rows', 500)


address_abbreviations = [
    "AL", "Alley", "AV", "Avenue", "BV", "Boulevard", "CA", "Canal", "CN", "Center",
    "CR", "Circle", "CT", "Court", "CV", "Cove", "DR", "Drive", "FW", "Freeway",
    "HY", "Highway", "LN", "Lane", "MO", "Motorway", "PY", "Parkway", "PA", "Path",
    "PL", "Place", "PZ", "Plaza", "RD", "Road", "SQ", "Square", "ST", "Street",
    "TE", "Terrace", "TR", "Trail", "WK", "Walk", "WY", "Way", "BL", "Broadway"
]

reg = re.compile('|'.join(address_abbreviations) + "$")
a = data_rel[~data_rel["LOCATION"].str.contains(reg)]["LOCATION"].unique() 

for val in a:
    print(val)
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="MADE-project")

for i in data_rel[(data_rel["LAT"] == 0) | (data_rel["LON"] == 0)].index:
    location = geolocator.geocode(data_rel["LOCATION"][i] + ", Los Angeles")
    if location:
        print(data_rel["LOCATION"][i] )
        print(location.address)
        print(location.latitude)
        print(location.longitude)
        print("###")
```
